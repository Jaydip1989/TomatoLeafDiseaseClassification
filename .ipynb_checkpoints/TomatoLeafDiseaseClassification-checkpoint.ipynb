```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.preprocessing.image import ImageDataGenerator, load_img, img_to_array
from tensorflow.keras.applications.inception_v3 import InceptionV3
from tensorflow.keras.applications.inception_v3 import preprocess_input
```

```python
IMAGE_SIZE = [256, 256]
batch_size  = 32
```

```python
from glob import glob
folders = glob('/Users/dipit/New Plant Diseases Dataset(Augmented)/train/*')
classes = len(folders)
classes
```

```python
def build_model():
    inception = InceptionV3(input_shape = (IMAGE_SIZE + [3]),
                           weights = 'imagenet',
                           include_top = False)
    x = layers.Flatten()(inception.output)
    prediction = layers.Dense(classes, activation="softmax")(x)
    model = keras.models.Model(inputs = inception.input, outputs = prediction)
    return model
```

```python
model = build_model()
model.summary()
```

```python
for layer in model.layers[:-1]:
    layer.trainable = False
```

```python
input_path = '/Users/dipit/New Plant Diseases Dataset(Augmented)/'
def preprocess_data(IMAGE_SIZE, batch_size):
    train_datagen = ImageDataGenerator(rescale = 1./255,
                                      shear_range = 0.2,
                                      zoom_range = 0.2,
                                      horizontal_flip = True)
    
    test_datagen = ImageDataGenerator(rescale = 1./255)
    
    train_generator = train_datagen.flow_from_directory(
        directory = input_path + 'train',
        target_size = IMAGE_SIZE,
        batch_size = batch_size,
        class_mode = 'categorical'
    )
    
    test_generator = test_datagen.flow_from_directory(
        directory = input_path + 'valid',
        target_size = IMAGE_SIZE,
        batch_size = batch_size,
        class_mode = 'categorical'
    )
    
    return train_generator, test_generator
```

```python
train_data, test_data = preprocess_data(IMAGE_SIZE, batch_size)
```

```python
model.compile(loss = 'categorical_crossentropy',
             optimizer = "adam",
             metrics = ['accuracy'])
epochs = 5
train_steps = len(train_data)
test_steps = len(test_data)
```

```python
print(train_steps, test_steps)
```

```python
history = model.fit(
    train_data, epochs = epochs,
    steps_per_epoch = train_steps,
    validation_data = test_data,
    validation_steps = test_steps
)
```

```python
history = model.fit(
    train_data, epochs = epochs,
    steps_per_epoch = train_steps,
    validation_data = test_data,
    validation_steps = test_steps
)
```

```python
model.save('TomatoLeafDiseaseInceptor.h5')
```

```python
scores = model.evaluate(test_data, verbose = 1)
print('Loss: ', scores[0])
print('Accuracy: ',scores[1])
```

```python
prediction = model.predict(test_data, verbose = 1)
y_pred = np.argmax(prediction, axis=1)
y_pred
```

```python
from sklearn.metrics import confusion_matrix, classification_report
class_labels = test_data.class_indices
class_labels = {v:k for k,v in class_labels.items()}
classes = list(class_labels.values())
```

```python
print("Confusion Matrix")
print(confusion_matrix(test_data.classes, y_pred))
```

```python
print(classification_report(test_data.classes, y_pred, target_names = classes))
```
